```python
import pandas as pd
import seaborn as sns
df=sns.load_dataset("tips")
df
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
#tips dataset there is no null values and duplicate values
```

```python
df
```

```python
d=sns.countplot(x="day",hue="sex",data=df)
df.groupby("day")["sex"].count()
#The number of customers is higher on weekends, especially Saturday and Sunday compared to weekdays. Saturday has the highest customer count overall.
```

```python
sns.countplot(x="sex",data=df)
df["sex"].value_counts()
#dataset contains more male customers compare to female customers
```

```python
sns.countplot(x="smoker",data=df)
df["smoker"].value_counts()
#dataset contains more non-smoked customers compare to smokers
```

```python
sns.countplot(x="day",data=df)
df["day"].value_counts()
#More customers visits at weekends compared to weekdays
```

```python
sns.countplot(x="time",data=df)
df["time"].value_counts()
#More customers visits at dinner time compared to lunch
```

```python
sns.countplot(x="size",data=df)
df["size"].value_counts()
#Most customers prefer sitting at tables of size 2 compared to other table sizes.
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex",data=df)
#there is a positive relation between total_bill and tip higher total_bill gives higher tips and female and male follows a similar pattern
```

```python
sns.boxplot(x="total_bill",data=df)
#it tells the range of total bills between 15 to 25 some customers are paid higher that outliers till 50
```

```python
sns.catplot(x="smoker",y="size",data=df,kind="strip")
#more number of smokers in 2 sized table followed 3 and 4,1 and 5 and there is no sokers in 6sized table
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
#There is a strong positive correlation between total_bill and tip, while size has a weak positive relationship with tip.
```

```python
sns.catplot(x="total_bill",y="tip",data=df,kind="strip")
#higher bill paid higher tips lower bill paid lower tips
```

```python
sns.violinplot(x="total_bill",data=df)
#sns.boxplot(x="total_bill",data=df)
#Most total bill values are concentrated in lower range with a few high-value outliers.
```

```python
sns.histplot(x="total_bill",data=df,kde=True)
df["total_bill"].value_counts()
#higher total bill distribution is at range 15 to 20
```

```python
sns.stripplot(x="sex",y="tip",data=df,jitter=True,dodge=True)
#tip amount distrbution of male and female customers with overlaping points
```

```python
sns.swarmplot(x="sex",y="tip",data=df,dodge=True)
#tip amount distrbution of male and female customers without overlaping points
```
